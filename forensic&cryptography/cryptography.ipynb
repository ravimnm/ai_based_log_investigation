```python
# ============================================================
# CRYPTOGRAPHIC FORENSIC MODULE
# Used for tamper-proof log evidence preservation
# ============================================================

import os
import json
import hashlib
from datetime import datetime, UTC

from cryptography.hazmat.primitives import hashes, serialization
from cryptography.hazmat.primitives.asymmetric import ec
from cryptography.hazmat.backends import default_backend


# ============================================================
# PATH CONFIGURATION
# ============================================================

BASE_PATH="/content/drive/MyDrive/Colab Notebooks/"

PRIVATE_KEY_PATH = BASE_PATH + "private_key.pem"
PUBLIC_KEY_PATH = BASE_PATH + "public_key.pem"


# ============================================================
# GENERATE KEYS IF NOT EXISTS
# ============================================================

if not os.path.exists(PRIVATE_KEY_PATH):

    print("Generating cryptographic keys...")

    private_key = ec.generate_private_key(
        ec.SECP256R1(),
        default_backend()
    )

    public_key = private_key.public_key()

    # Save private key
    with open(PRIVATE_KEY_PATH, "wb") as f:
        f.write(
            private_key.private_bytes(
                encoding=serialization.Encoding.PEM,
                format=serialization.PrivateFormat.PKCS8,
                encryption_algorithm=serialization.NoEncryption()
            )
        )

    # Save public key
    with open(PUBLIC_KEY_PATH, "wb") as f:
        f.write(
            public_key.public_bytes(
                encoding=serialization.Encoding.PEM,
                format=serialization.PublicFormat.SubjectPublicKeyInfo
            )
        )

    print("Keys generated.")

# ============================================================
# LOAD KEYS
# ============================================================

with open(PRIVATE_KEY_PATH, "rb") as f:
    private_key = serialization.load_pem_private_key(
        f.read(),
        password=None,
        backend=default_backend()
    )

with open(PUBLIC_KEY_PATH, "rb") as f:
    public_key = serialization.load_pem_public_key(
        f.read(),
        backend=default_backend()
    )

print("🔐 Cryptographic keys ready.")


# ============================================================
# FORENSIC EVIDENCE FREEZE
# ============================================================

def freeze_log_structure(log_dict, risk_score, model_version, log_source):

    frozen = {

        "log_source": log_source,
        "original_log": log_dict,
        "risk_score": float(risk_score),
        "model_version": model_version,
        "decision_timestamp": datetime.now(UTC).isoformat()

    }

    return frozen


# ============================================================
# HASH GENERATION (EVIDENCE INTEGRITY)
# ============================================================

def generate_hash(data):

    serialized = json.dumps(data, sort_keys=True).encode()

    hash_value = hashlib.sha256(serialized).hexdigest()

    return hash_value


# ============================================================
# SIGN HASH (DIGITAL SIGNATURE)
# ============================================================

def sign_hash(hash_value):

    signature = private_key.sign(

        hash_value.encode(),
        ec.ECDSA(hashes.SHA256())

    )

    return signature.hex()


# ============================================================
# VERIFY SIGNATURE
# ============================================================

def verify_signature(hash_value, signature_hex):

    try:

        signature = bytes.fromhex(signature_hex)

        public_key.verify(

            signature,
            hash_value.encode(),
            ec.ECDSA(hashes.SHA256())

        )

        return True

    except Exception:

        return False


# ============================================================
# COMPLETE FORENSIC EVIDENCE CREATION
# ============================================================

def create_signed_evidence(log_dict, risk_score, model_version, log_source):

    frozen = freeze_log_structure(
        log_dict,
        risk_score,
        model_version,
        log_source
    )

    hash_value = generate_hash(frozen)

    signature = sign_hash(hash_value)

    frozen["evidence_hash"] = hash_value
    frozen["signature"] = signature

    return frozen


print("Forensic cryptography module loaded.")
```
